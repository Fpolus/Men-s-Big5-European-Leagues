```python
# Import Dependencies
from splinter import Browser
from bs4 import BeautifulSoup
from pprint import pprint

import pandas as pd
```

```python
# Create a Chrome browser instance
browser = Browser('chrome')
```

```python
# Visit the page and create the soup object
url = 'https://fbref.com/en/comps/9/Premier-League-Stats'
browser.visit(url)

# Extract objects from the webpage
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
tables = soup.find_all('tbody')

pprint(tables)
```

```python
# pulling all teams from the url above
target_table = soup.find_all('table')[0]  

# Create empty lists to store team names, competition names, and links
team_list = []
link_list = []

# Iterate through the rows in the target table
for row in target_table.find_all('tr'):
    # Find the first <th> or <td> cell with the class "left". Here both the team name and competition name have been found
    td_cell = row.find('td', {'class': 'left'})
    
    if td_cell:
        # Find the <a> tag within the cell
        link_cell = td_cell.find('a')
        
        if link_cell:
            # Extract the href and link text
            href = link_cell['href']
            link_text = link_cell.get_text()
            
            # Append the href and link text to the lists
            team_list.append(link_text)
            link_list.append(href)
            
# Create DataFrame from the extracted data
df = pd.DataFrame({'team': team_list, 'link': link_list})

# Save the DataFrame to a CSV file
df.to_csv('Resources/Premier_league_teams.csv', index=False)

```
