```python
# Import Dependencies
from splinter import Browser
from bs4 import BeautifulSoup
from pprint import pprint

import pandas as pd
```

```python
# Create a Chrome browser instance
browser = Browser('chrome')
```

```python
# Visit the page and create the soup object
url = 'https://fbref.com/en/comps/9/Premier-League-Stats'
browser.visit(url)

# Extract objects from the webpage
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
tables = soup.find_all('tbody')

pprint(tables)
```

```python
# Iterate over the years 2017 to 2023
for year in range(2017, 2024):
    url = f'https://fbref.com/en/comps/9/{year}-{year+1}/{year}-{year+1}-Premier-League-Stats'
    browser.visit(url)
    
    # Extract objects from the webpage
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    target_table = soup.find_all('table')[0]
    
    # Create empty lists to store team names and links
    team_list = []
    link_list = []
    
    # Iterate through the rows in the target table
    for row in target_table.find_all('tr'):
        td_cell = row.find('td', {'class': 'left'})
        
        if td_cell:
            link_cell = td_cell.find('a')
            
            if link_cell:
                href = link_cell['href']
                link_text = link_cell.get_text()
                team_list.append(link_text)
                link_list.append(href)

    # Create DataFrame from the extracted data
    df = pd.DataFrame({'team': team_list, 'link': link_list})

    # Save the DataFrame to a CSV file for the current year
    file_name = f'../Resources/premier_league_teams_{year}_{year+1}.csv'
    df.to_csv(file_name, index=False)

# Close the browser when done
browser.quit()
```
